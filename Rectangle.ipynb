Load the **images**


```
img_size = 50
def load_image(file):
    img = image.img_to_array(image.load_img(file,target_size=(img_size,img_size),color_mode="grayscale"))
    return np.reshape(img,(img_size,img_size))
```

```
train_images = [] 
train_labels = []

for i in range(10000):
    img = load_image("dataset/Res/" + str(i) + ".jpg")
    train_images.append(img)

    mask = load_image("dataset/Res/" + str(i) + ".png")
    train_labels.append(np.reshape(mask,(img_size*img_size)))

train_labels = np.array(train_labels)
train_images = np.array(train_images)
train_images=train_images/255
train_labels=train_labels/255
```

Create the **model**

```
model = Sequential([
    Flatten(input_shape=(img_size,img_size)),
    Dense(units=img_size*img_size,activation='sigmoid')
])
model.compile(optimizer=Adam(learning_rate=0.0001),loss='binary_crossentropy',metrics=['accuracy'])
```

Train the **model**

```
model.fit(train_images,train_labels,epochs=1000)
```

Test the **model** with an image from the test set

```
img_prediction = load_image("dataset/Test/2.jpg")
img_prediction = img_prediction/255
img_prediction = np.reshape(img_prediction,(1,img_size,img_size))

prediction = model.predict(img_prediction)
prediction = prediction*225
prediction = np.reshape(prediction,(img_size,img_size,1))
img = image.array_to_img(prediction)

plt.figure()
plt.imshow(img)
plt.show()
```
